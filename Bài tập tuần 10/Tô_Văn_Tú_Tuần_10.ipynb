```python
import pandas as pd
import numpy as np
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import itertools

data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/timeseries/AirPassengers.csv')
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

ts = data['#Passengers']

train_size = int(len(ts) * 0.8)
train, test = ts[0:train_size], ts[train_size:]

best_aic_ar = np.inf
best_order_ar = None
best_model_ar = None

for lag in range(1, 20):
    try:
        model = AutoReg(train, lags=lag)
        results = model.fit()
        if results.aic < best_aic_ar:
            best_aic_ar = results.aic
            best_order_ar = lag
            best_model_ar = results
    except:
        continue

print(f'Best AR model order: {best_order_ar}, AIC: {best_aic_ar}')

# Ti u tham s cho MA model
best_aic_ma = np.inf
best_order_ma = None
best_model_ma = None

for q in range(1, 20):
    try:
        model = ARIMA(train, order=(0, 0, q))
        results = model.fit()
        if results.aic < best_aic_ma:
            best_aic_ma = results.aic
            best_order_ma = q
            best_model_ma = results
    except:
        continue

print(f'Best MA model order: {best_order_ma}, AIC: {best_aic_ma}')

# Ti u tham s cho ARMA model
best_aic_arma = np.inf
best_order_arma = None
best_model_arma = None

for p, q in itertools.product(range(1, 5), range(1, 5)):
    try:
        model = ARIMA(train, order=(p, 0, q))
        results = model.fit()
        if results.aic < best_aic_arma:
            best_aic_arma = results.aic
            best_order_arma = (p, q)
            best_model_arma = results
    except:
        continue

print(f'Best ARMA model order: {best_order_arma}, AIC: {best_aic_arma}')

# Ti u tham s cho ARIMA model
best_aic_arima = np.inf
best_order_arima = None
best_model_arima = None

for p, d, q in itertools.product(range(1, 5), range(0, 3), range(1, 5)):
    try:
        model = ARIMA(train, order=(p, d, q))
        results = model.fit()
        if results.aic < best_aic_arima:
            best_aic_arima = results.aic
            best_order_arima = (p, d, q)
            best_model_arima = results
    except:
        continue

print(f'Best ARIMA model order: {best_order_arima}, AIC: {best_aic_arima}')

# nh gi m hnh
def evaluate_model(model, test):
    predictions = model.forecast(steps=len(test))
    rmse = np.sqrt(mean_squared_error(test, predictions))
    return rmse

# Kim tra v nh gi cc m hnh
if best_model_ar is not None:
    ar_rmse = evaluate_model(best_model_ar, test)
else:
    ar_rmse = np.inf

if best_model_ma is not None:
    ma_rmse = evaluate_model(best_model_ma, test)
else:
    ma_rmse = np.inf

if best_model_arma is not None:
    arma_rmse = evaluate_model(best_model_arma, test)
else:
    arma_rmse = np.inf

if best_model_arima is not None:
    arima_rmse = evaluate_model(best_model_arima, test)
else:
    arima_rmse = np.inf

print(f'AR model RMSE: {ar_rmse}')
print(f'MA model RMSE: {ma_rmse}')
print(f'ARMA model RMSE: {arma_rmse}')
print(f'ARIMA model RMSE: {arima_rmse}')

models_rmse = {
    'AR': ar_rmse,
    'MA': ma_rmse,
    'ARMA': arma_rmse,
    'ARIMA': arima_rmse
}

best_model_name = min(models_rmse, key=models_rmse.get)
print(f'The best model based on RMSE is: {best_model_name}')
```
